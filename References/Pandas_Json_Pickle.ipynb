### save df to csv pandas

```python
import seaborn as sns

iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.to_csv('/content/iris.csv')
```

### operations

##### random selection

```python
iris.sample(frac=0.5)
```

#### get n largest/smallest values

```python
iris.nlargest(10, 'petal_length')
```

```python
iris.nsmallest(10, 'petal_length')
```

#### iloc/loc

```python
# 10 rows
iris.iloc[10:20]
```

```python
# 11 rows
iris.loc[10:20]
```

```python
# fetch columns with range
iris.loc[:,'sepal_width':'petal_length']
```

```python
#fetch columns with positions
iris.iloc[:,[2,3,4]]
```

```python
# fetch colums with name
iris[['petal_length', 'petal_width', 'species']]
```

```python
# fetch rows with condition and columns with names
iris.loc[iris['petal_length'] > 3, ['petal_width','species']]
```

#### Rename

```python
s_iris = iris.rename(columns = {'sepal_width':'s'})
s_iris
```

#### Drop

```python
s_iris = s_iris.drop(columns=['s','petal_length'])
s_iris
```

#### Apply func to column

```python
def add1(num):
  return num+1
s_iris['new'] = s_iris.petal_width.apply(add1)
```

# Json

```python
import json
```

```python
iris
```

#### saving to json

```python
iris_dict = iris.to_dict()
print(iris_dict)
with open('/content/iris.json',"w") as file:
    json.dump(iris_dict, file, indent=4)
    print("Saved File")
```

#### loading json data

```python
with open("/content/iris.json") as file:
    jsonData = json.load(file)
print(jsonData)
```
