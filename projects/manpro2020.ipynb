```
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
from mpl_toolkits import mplot3d
from scipy.special import kn
```

```
x = np.arange(-301,301)
y = np.arange(-301,301)
X,Y = np.meshgrid(x,y)
```

```
def new_profile(Z,x,y,X,Y,i=0,inertia=0.5,std=0.02,t0=10000):
  # inertia=0.5
  # std=0.02
  # t0=10000
  z_new=inertia*Z +  t0*np.exp(-np.sqrt((X-x)**2+(Y-y)**2)*std)
  if(i==0):
    plt.figure(figsize=(8,8))
    plt.imshow(z_new, cmap='rainbow')
    plt.contour(z_new, 1, colors='cyan', linewidths=2)

  if(i==1):
    plt.figure(figsize=(8,8))
    plt.contour(X, Y, z_new, 10, cmap='RdGy')
    # plt.colorbar();
  if(i==2):
    fig = plt.figure(figsize=(8,8))
    ax = plt.axes(projection='3d')
    ax.contour3D(X, Y, z_new, 100, cmap='rainbow')

  return z_new
z=np.zeros(X.shape)+80

z1=new_profile(z,0,0,X,Y)
```
